## Data collection

### Subtask:
Describe the dataset and its source, which is provided as `/content/heat_recovery_anomaly_detection_dataset.csv`.


**Reasoning**:
Load the dataset, display the first few rows, print the shape, and display data types to start describing the dataset.



```python
import pandas as pd

file_path = '/content/heat_recovery_anomaly_detection_dataset.csv'
df = pd.read_csv(file_path)

print("First 5 rows of the dataset:")
display(df.head())

print("\nShape of the dataset (rows, columns):")
print(df.shape)

print("\nData types of each column:")
print(df.info())

print("\nDataset source:")
print(file_path)
```

## Data cleaning

### Subtask:
Identify and handle missing values, outliers, and inconsistencies in the dataset.


**Reasoning**:
Check for missing values, convert the timestamp column to datetime objects, and calculate descriptive statistics for numerical columns to identify potential outliers.



```python
# 1. Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# 3. Convert the 'timestamp' column to a datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 2. Identify potential outliers using descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df.describe())

# 4. Report findings on missing values and potential outliers.
# Based on the output of df.isnull().sum(), we can see if there are any missing values.
# Based on the output of df.describe(), we can observe the range (min and max)
# and distribution (mean, std, quartiles) of the numerical features, which can
# indicate the presence of potential outliers (e.g., values significantly far from the mean or quartiles).
# We will report these findings based on the printed outputs.
```

## Data transformation

### Subtask:
Perform necessary data transformations, such as scaling or encoding, to prepare the data for modeling.


**Reasoning**:
Separate the timestamp column and apply standard scaling to the numerical features.



```python
from sklearn.preprocessing import StandardScaler

# Separate the timestamp column
timestamp_column = df['timestamp']
numerical_features = df.drop('timestamp', axis=1)

# Apply a standard scaler to the numerical features
scaler = StandardScaler()
scaled_numerical_features = scaler.fit_transform(numerical_features)

# Store the scaled numerical features in a new DataFrame
scaled_df = pd.DataFrame(scaled_numerical_features, columns=numerical_features.columns)

# Display the first few rows of the scaled DataFrame
display(scaled_df.head())
```

## Feature engineering

### Subtask:
Create new features from existing ones that could improve the performance of the anomaly detection model.


**Reasoning**:
Extract time-based features, calculate rolling averages, and append them to the scaled DataFrame.



```python
# 1. Extract time-based features from the 'timestamp' column
scaled_df['hour'] = timestamp_column.dt.hour
scaled_df['dayofweek'] = timestamp_column.dt.dayofweek
scaled_df['month'] = timestamp_column.dt.month

# 2. Calculate rolling averages for relevant numerical columns
# We will calculate a rolling average with a window of 60 (representing 1 hour) for all numerical columns
window_size = 60
rolling_avg_df = scaled_df.rolling(window=window_size).mean()

# Rename the columns to indicate they are rolling averages
rolling_avg_df.columns = [f'{col}_rolling_avg' for col in rolling_avg_df.columns]

# Drop the time-based features columns from the rolling average dataframe before joining
rolling_avg_df = rolling_avg_df.drop(columns=['hour_rolling_avg', 'dayofweek_rolling_avg', 'month_rolling_avg'])

# 3. Consider creating interaction terms or polynomial features (Skipped for simplicity in this step, can be added later based on model performance)

# 4. Append the newly engineered features (rolling averages) to the existing scaled DataFrame
scaled_df = pd.concat([scaled_df, rolling_avg_df], axis=1)

# Drop rows with NaN values resulting from the rolling average calculation
scaled_df.dropna(inplace=True)

# Reset index after dropping rows
scaled_df.reset_index(drop=True, inplace=True)

# 5. Display the first few rows and the columns of the updated DataFrame
print("First 5 rows of the updated DataFrame with new features:")
display(scaled_df.head())

print("\nColumns of the updated DataFrame:")
print(scaled_df.columns)
```

## Model selection

### Subtask:
Choose appropriate anomaly detection algorithms based on the nature of the data and the problem.


**Reasoning**:
Based on the problem description and the nature of the dataset, I will choose Isolation Forest and Local Outlier Factor (LOF) as suitable anomaly detection algorithms and briefly justify the choices and mention their characteristics.



```python
# 1. Identify at least two suitable anomaly detection algorithms.
# Isolation Forest is suitable for large datasets and can handle high-dimensional data.
# LOF is a density-based algorithm that is effective in identifying local outliers.
# Both are unsupervised learning algorithms, which fits the problem of anomaly detection where we don't have labeled anomalies.

print("Chosen Anomaly Detection Algorithms:")
print("- Isolation Forest")
print("- Local Outlier Factor (LOF)")

# 2. Briefly justify the choice of each algorithm.
print("\nJustification for choosing Isolation Forest:")
print("Isolation Forest is an ensemble tree-based anomaly detection method. It is effective because it explicitly isolates anomalies instead of profiling normal data points. This makes it computationally efficient and capable of handling high-dimensional datasets. It is particularly well-suited for identifying outliers in large datasets like this one.")

print("\nJustification for choosing Local Outlier Factor (LOF):")
print("Local Outlier Factor (LOF) is a density-based anomaly detection algorithm that measures the local deviation of a given data point with respect to its neighbors. It is effective at finding outliers that are not necessarily global outliers but are anomalous with respect to their local neighborhood. This can be useful for detecting different types of anomalies in the heat recovery system data.")

# 3. Mention the key characteristics of each selected algorithm.
print("\nKey Characteristics of Isolation Forest:")
print("- Tree-based: It builds isolation trees to isolate anomalies.")
print("- Unsupervised: It does not require labeled data.")
print("- Efficient for large datasets: Its complexity is linear with respect to the number of instances.")
print("- Handles high-dimensional data: It works well in high-dimensional spaces.")

print("\nKey Characteristics of Local Outlier Factor (LOF):")
print("- Density-based: It identifies outliers based on local density deviations.")
print("- Unsupervised: It does not require labeled data.")
print("- Effective for local outliers: It can find anomalies that are only outliers relative to their neighbors.")
print("- Requires neighborhood size parameter (k): The performance can depend on the choice of k.")
```

## Training

### Subtask:
Train the selected anomaly detection models (Isolation Forest and LOF) on the prepared data.


**Reasoning**:
Train the selected anomaly detection models (Isolation Forest and LOF) on the prepared data and obtain the anomaly scores or labels.



```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

# Instantiate Isolation Forest model
isolation_forest = IsolationForest(random_state=42)

# Instantiate Local Outlier Factor model
# novelty=False is used when the data for fitting is the same as the data for predicting
lof = LocalOutlierFactor(novelty=False)

# Train Isolation Forest model
isolation_forest.fit(scaled_df)

# Train Local Outlier Factor model and get anomaly labels
lof_anomaly_labels = lof.fit_predict(scaled_df)

# Get anomaly scores from Isolation Forest
isolation_forest_scores = isolation_forest.decision_function(scaled_df)

# Store the anomaly scores and labels
scaled_df['isolation_forest_score'] = isolation_forest_scores
scaled_df['lof_anomaly_label'] = lof_anomaly_labels

print("Anomaly detection models trained and scores/labels obtained.")
display(scaled_df.head())
```

## Evaluation

### Subtask:
Evaluate the performance of the trained anomaly detection models (Isolation Forest and LOF) using appropriate metrics.


**Reasoning**:
Evaluate the performance of the trained anomaly detection models by examining the distribution of Isolation Forest scores and the count/percentage of anomalies detected by LOF.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Examine the distribution of isolation_forest_score
print("Descriptive statistics for Isolation Forest scores:")
display(scaled_df['isolation_forest_score'].describe())

plt.figure(figsize=(10, 6))
sns.histplot(scaled_df['isolation_forest_score'], bins=50, kde=True)
plt.title('Distribution of Isolation Forest Anomaly Scores')
plt.xlabel('Anomaly Score')
plt.ylabel('Frequency')
plt.show()

# 3. Calculate the percentage/count of anomalies detected by LOF
lof_anomaly_count = scaled_df[scaled_df['lof_anomaly_label'] == -1].shape[0]
total_count = scaled_df.shape[0]
lof_anomaly_percentage = (lof_anomaly_count / total_count) * 100

print(f"\nTotal number of data points: {total_count}")
print(f"Number of anomalies detected by LOF: {lof_anomaly_count}")
print(f"Percentage of anomalies detected by LOF: {lof_anomaly_percentage:.2f}%")

# 5. Discuss limitations of evaluation without true labels
print("\nLimitations of evaluating unsupervised anomaly detection models without true labels:")
print("Evaluating unsupervised anomaly detection models like Isolation Forest and LOF without true anomaly labels is challenging.")
print("Traditional supervised learning metrics such as precision, recall, F1-score, and accuracy cannot be directly applied.")
print("Evaluation relies heavily on:")
print("- Examining the distribution of anomaly scores (e.g., Isolation Forest scores) to see how well anomalies are separated from normal data.")
print("- Analyzing the number or percentage of data points flagged as anomalies based on model thresholds or labels (e.g., LOF labels).")
print("- Visual inspection of the data, potentially in reduced dimensions, although this can be complex for high-dimensional data.")
print("The interpretation of results is more subjective and depends on domain knowledge to validate if the detected anomalies are meaningful.")
```

## Parameter tuning

### Subtask:
Optimize the parameters of the chosen model(s) (Isolation Forest and LOF) to improve performance.


**Reasoning**:
Experiment with different contamination values for Isolation Forest and different n_neighbors values for LOF to see how they affect the number of detected anomalies.



```python
# Experiment with Isolation Forest contamination parameter
contamination_values_if = [0.01, 0.05, 0.1]
print("Experimenting with Isolation Forest contamination parameter:")
for contamination in contamination_values_if:
    isolation_forest_exp = IsolationForest(contamination=contamination, random_state=42)
    isolation_forest_exp.fit(scaled_df.drop(['isolation_forest_score', 'lof_anomaly_label'], axis=1))
    isolation_forest_labels_exp = isolation_forest_exp.predict(scaled_df.drop(['isolation_forest_score', 'lof_anomaly_label'], axis=1))
    n_anomalies_if = (isolation_forest_labels_exp == -1).sum()
    print(f"Contamination: {contamination}, Detected anomalies: {n_anomalies_if}")

# Experiment with LOF n_neighbors parameter
n_neighbors_values_lof = [20, 50, 100]
print("\nExperimenting with LOF n_neighbors parameter:")
for n_neighbors in n_neighbors_values_lof:
    lof_exp = LocalOutlierFactor(n_neighbors=n_neighbors, novelty=False)
    lof_labels_exp = lof_exp.fit_predict(scaled_df.drop(['isolation_forest_score', 'lof_anomaly_label'], axis=1))
    n_anomalies_lof = (lof_labels_exp == -1).sum()
    print(f"n_neighbors: {n_neighbors}, Detected anomalies: {n_anomalies_lof}")
```

**Reasoning**:
Based on the experimental results, summarize the findings for each model and select the parameters that seem most reasonable.



```python
# Summarize findings and select parameters

print("Summary of Parameter Experimentation:")

print("\nIsolation Forest:")
print("Varying the 'contamination' parameter directly influences the number of detected anomalies. A contamination of 0.01 detected 800 anomalies, 0.05 detected 3997, and 0.1 detected 7994.")
print("Without true labels, selecting the optimal contamination is subjective. A value of 0.05 or 0.1 might be more reasonable if we expect a higher percentage of anomalies in the system, but 0.01 might be too restrictive.")
selected_contamination = 0.05
print(f"Selected contamination for Isolation Forest: {selected_contamination} (This choice is based on a hypothetical expectation of anomaly percentage; further validation would require domain expertise or labeled data.)")

print("\nLocal Outlier Factor (LOF):")
print("Experimenting with 'n_neighbors' (20, 50, 100) did not result in the detection of any anomalies with the current dataset and scaling. This could indicate that the anomalies, if they exist, are not local density-based outliers that LOF is designed to detect, or that the chosen range of n_neighbors is not appropriate, or that the dataset does not contain clear anomalies detectable by LOF with these parameters.")
print("Given that no anomalies were detected with the tested n_neighbors values, it's difficult to select an optimal parameter for LOF based on this experiment alone. Further investigation or different approaches might be needed for LOF.")
selected_n_neighbors = 100 # Keep the largest tested value for potential future exploration
print(f"Selected n_neighbors for LOF: {selected_n_neighbors} (No anomalies were detected with tested values; this is a placeholder for potential further exploration.)")

# Note: The selection of parameters without true labels is based on observing the number of detected anomalies
# and making an informed guess about the expected anomaly rate in the system.
# In a real-world scenario, this would involve domain expertise or a small set of labeled data for validation.
```
